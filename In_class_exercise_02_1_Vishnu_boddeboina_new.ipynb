```python
# You code here (Please add comments in the code):

# import libraries
from bs4 import BeautifulSoup
import requests
import urllib.request
import re
import pandas as pd


# headers to open the link

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',
    'Accept-Language': 'en-US, en;q=0.5'
}


r_titles = []
r_ratings = []
r_text=[]
r_dates=[]
for pg_num in range(1,150):
  r_url = f'https://www.amazon.com/Zacurate-Fingertip-Oximeter-Saturation-Batteries/product-reviews/B07PQ8WTC4/ref=cm_cr_arp_d_paging_btm_next_{str(pg_num)}?ie=UTF8&reviewerType=all_reviews&pageNumber={str(pg_num)}'
 
  # Launch the page
  page = requests.get(r_url, headers=headers)
  # read page content
  soup = BeautifulSoup(page.content,'html.parser')

  # Select all title elemets
  review_titles = soup.find_all('a',class_='review-title-content')
  
  # Extract title text from the dom element
  for t in range(len(review_titles)):
    r_titles.append(review_titles[t].get_text().strip())

  # Select all rating dom elements
  rating_dom = soup.find_all('i',class_='review-rating')
  
  # Extract ratings from the dom element
  for r in range(2,len(rating_dom)):
    r_ratings.append(rating_dom[r].get_text().strip())  
  
   # Select all rating dom elements
  review_text_dom= soup.find_all('span',class_='review-text-content')
  
  # Extract review text text from the dom element
  for rt in range(0,len(review_text_dom)):
    r_text.append(review_text_dom[rt].get_text().strip())


# Create a data frame 
df_reviews = pd.DataFrame({
    'Title' : r_titles,
    'Rating': r_ratings,
    'Review': r_text
})
print(f'Shape of the data frame: {df_reviews.shape}')
df_reviews
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2012-2022).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):
from bs4 import BeautifulSoup
import requests
import urllib.request 
import re
import pandas as pd
df_articles = pd.DataFrame()


# headers to open the link

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',
    'Accept-Language': 'en-US, en;q=0.5'
}
for pg_num in range(1, 20):
  # URL with search string
  url = f'https://scholar.google.com/scholar?start={str(pg_num)}&q=information+retrieval&hl=en&as_sdt=0,44&as_ylo=2012&as_yhi=2022'
  pg = requests.get(url, headers=headers)

  # Read DOM using BeautifulSoup
  soup = BeautifulSoup(pg.content,'html.parser')

  # Select each article from search results
  for article in soup.select('[data-lid]'):
    # get title text
    title = article.select_one('.gs_rt').text
    # get journal details 
    journal = article.select_one('.gs_a').text

    # get publisher details 
    publisher= article.select('a')[0].get_text()
  
    pub_info = article.find_all('div',class_='gs_a')[0].get_text()
    
    # extract authors
    authors= pub_info.split('-')[0]
    abstract= article.find_all('div',class_='gs_rs')[0].get_text()

    # Extract year from publisher info
    b_year = re.findall(r'\d+', pub_info)
    
    info = list(map(int, b_year))
    year = str(info)[1:-1]

    # Add data to data frame
    df_articles = df_articles.append(pd.DataFrame({'Title': title, 'Journal': journal, 'Publisher':publisher, 'Year': year, 'Authors': authors, 'Abstract': abstract}, index=[0]))

df_articles.head()
```
